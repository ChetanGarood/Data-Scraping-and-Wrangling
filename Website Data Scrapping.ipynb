```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
url='https://www.oyorooms.com/hotels-in-coimbatore/?page={}'
browser=webdriver.Firefox()
oyomaster={}
i=1
while(True):
    
    
    url1=url.format(i)
    browser.get(url1)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    rows=soup.find_all('div',class_='hotelCardListing__descriptionWrapper')
    if(len(rows)==0):
        break
    pri=[]
    add=[]
    names=[]
    for row in rows:
        prices=row.find_all('span',class_='listingPrice__finalPrice')
        if len(prices)==0:
            pri.append('Doesnt Exist')
        else:
            pri.append(prices[0].text)
        adress=row.find_all('span',itemprop="streetAddress")
        if len(adress)==0:
            add.append('Doesnt Exist')
        else:
            add.append(adress[0].text)
        name=row.find_all('h3',itemprop="name")
        if len(name)==0:
            names.append('Doesnt Exist')
        else:
            names.append(name[0].text)
        table=pd.DataFrame({'Name':names,'Address':add,'Price':pri})
        oyomaster[i]=table
    i=i+1
oyo=pd.DataFrame()
for i in oyomaster.keys():
    oyo=pd.concat((oyo,oyomaster[i]),axis=0)
```

```python
all_rew=browser.find_element_by_xpath("/html/body/div[7]/div[1]/div/div[1]/div[3]/div/div[3]/div[1]/div[1]/div[1]/div/div[3]/a[2]")
all_rew.click()
zreds=browser.find_elements_by_class_name("zred")
len(zreds)
```

```python
while True:
    zreds=browser.find_elements_by_class_name("zred")
    for zred in zreds:
        if zred.text.strip().lower()=="load more":
            zred.click()
```

```python
import time
flag=True
while flag:
    state={}
    zreds=browser.find_elements_by_class_name('zred')
    for zred in zreds:
        if zred.text.strip().lower()=='load more':
            zred.click()
            state[zred]=1
            time.sleep(0.7)
        else:
            state[zred]=0
        if sum(state.values())==1:
            flag=True
        else:
            flag=False
```
